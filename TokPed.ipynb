```python
import time
import openpyxl
from selenium import webdriver 
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
from pandas import ExcelWriter
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
s=Service(ChromeDriverManager().install())
url = "https://www.tokopedia.com/search?st=product&q=elektronik&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&navsource="
driver = webdriver.Chrome(service=s)
driver.get(url)

WebDriverWait(driver,5).until(EC.presence_of_element_located((By.CSS_SELECTOR,"#zeus-root")))
time.sleep(2)

for i in range(25):
    driver.execute_script("window.scrollBy(0,250)")
    time.sleep(1) 

soup = BeautifulSoup(driver.page_source,"html.parser")
```

```python
key = 0
for item in soup.findAll('div',class_="css-974ipl"):
    product_name = item.find('div',class_="prd_link-product-name css-svipq6").text
    price = item.find('div',class_='prd_link-product-price css-1ksb19c').text
    ratings = item.find('span', class_= "prd_rating-average-text css-t70v7i")
    
    if bool(ratings) == True :
        rating = ratings.text
    else :
        rating = "has no rating"
    
    solds = item.find('span',class_='prd_label-integrity css-1duhs3e')
    if bool(solds) == True :
        sold = solds.text
    else :
        sold = "has no soldout"
    
    discounts = item.find('div',class_="prd_badge-product-discount css-1qtulwh")
    if bool(discounts) == True :
        discount = discounts.text
    else :
        discount = "Dont discount"
    
    starting_prices = item.find('div', class_="prd_label-product-slash-price css-1u1z2kp")
    if bool(starting_prices) == True :
        starting_price = starting_prices.text
    else :
        starting_price = "dont have starting price"

    for locate in item.findAll('div',class_='css-1rn0irl'):
        shop_name = locate.find('span',class_='prd_link-shop-name css-1kdc32b flip').text
        
    key += 1
    print(key,product_name,shop_name,rating,sold,discount,starting_price)
    # print(starting_price)
    df = pd.DataFrame([[product_name,price,shop_name,rating,sold]],index=[key],columns=['a','b','c','d','e'])
    # print (df)
```
